```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Data Frame of Data

```python
df=pd.read_csv("sub-division_population_of_pakistan.csv")
```

# Males ,Feamles,Transgenders (Rural Areas )

```python
def No_counts(males,females,trans=None):
    Males_Rulars=df[males].sum()
    Females_Rulars=df[females].sum()
    if trans:
        Transgender_Rulars=df[trans].sum() 
        x_lables=[males,females,trans]
        y_lables=[Males_Rulars, Females_Rulars,Transgender_Rulars]
        
    else:
         
        x_lables=[males,females]
        y_lables=[Males_Rulars, Females_Rulars]
      
        
    plt.figure(figsize=(6, 6))
    bar_plots=sns.barplot(x=x_lables, y=y_lables ,palette="dark", hue=x_lables, legend=False) 
    plt.title(f"No of {males} {females} {trans}") 
 
    for bar in bar_plots.containers:
        bar_plots.bar_label(bar,fmt="%.0f")
        
    
    plt.show()


No_counts('MALE (RURAL)', 'FEMALE (RURAL)','TRANSGENDER (RURAL)')
```

#### From Above i anlayze that the population of males is greater then the Females and transgenders in rural areas 

#  Males ,Females,Transgenders (Urbans Areas )

```python
No_counts("MALE (URBAN)","FEMALE (URBAN)","TRANSGENDER (URBAN)")
```

#### From Above Charts We analyze that the Males is more then then the Females and Transgenders in Urban Areas

# On The Basis of  Province Popltion CountOf Rural Areas

```python
all_provicnes=df["PROVINCE"].unique() 
```

```python
def Pie_charts(a,colors=None):
  result_data2= []
  labels=[]

  for province in all_provinces:
        province_check=df[df["PROVINCE"]==province]
    # print(province_check)
        sum_of_data=province_check[a].sum()
        result_data2.append(sum_of_data)
        labels.append(province)

  plt.figure(figsize=(7,10))
  plt.pie(result_data2,labels=labels, autopct='%1.2f%%',shadow=True,colors=colors)
  plt.title(f"Province Population of {a} areas")   
  plt.show()



colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']


Pie_charts("POPULATION 1998 (RURAL)",  colors)

 



```

#### From Above Chart I Analyze that the No The population of Rural Area is Effcted by the Province

## Conclusion

### Gender Distribution:
The data indicates a higher number of males compared to females and transgenders in both rural and urban areas. This trend is consistent across different provinces, suggesting a gender imbalance that might need further investigation to understand the underlying causes.

### Provincial Population Impact:
Both rural and urban populations vary significantly across provinces. This variation indicates that different provinces have distinct demographic profiles, influenced by factors such as economic opportunities, cultural norms, and migration patterns.

### Rural vs. Urban Population:
The rural population exceeds the urban population in the regions studied. This might reflect the agricultural and traditional livelihoods predominant in these areas, where large rural communities are sustained by farming and related activities.

### Annual Growth Rate:
The higher annual growth rate in rural areas compared to urban areas suggests that rural regions are experiencing faster population growth. This could be due to higher birth rates, lower mortality rates, or rural-to-urban migration trends that might not be as pronounced as expected.

### Household Size:
Larger household sizes in rural areas indicate extended family living arrangements, which are more common in rural settings compared to urban areas where nuclear families might be more prevalent due to different economic and social dynamics.
