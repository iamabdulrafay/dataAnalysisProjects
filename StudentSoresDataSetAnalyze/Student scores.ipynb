```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Data Frame UnamedCharts removed

# SeaBorn Count Plot Creation

```python
sns.set_palette("dark")
count_plot=sns.countplot(data=df,x="Gender")
count_plot.bar_label(count_plot.containers[0]) 
plt.title("Student Genders")
plt.show()
```

#### **From the above charts we obeserve the no of female is more than the no of males**

# Find the relation between the student marks on the basis of Parents Education

```python
parent_edu_group=df.groupby("ParentEduc").agg({
    
 "MathScore":"mean",	"ReadingScore":"mean",	"WritingScore":"mean"
    
})


```

# student score charts on the basis of Parential Edu

```python
sns.heatmap(data=parent_edu_group,annot=True, cmap="YlGnBu") 
plt.show()
```

####  From the Above anaylisis we observe the Impact of parent Eduction on the students Score

# student score charts on the basis of Parent Martial Status

```python
parent_maritial_group=df.groupby("ParentMaritalStatus").agg({ "MathScore":"mean",	"ReadingScore":"mean",	"WritingScore":"mean"})
```

```python
parent_maritial_group
```

```python
sns.heatmap(data=parent_maritial_group,annot=True)
plt.show()
```

# student score charts on the basis of Weekly Study Hours

```python
weekly_study_group=df.groupby("WklyStudyHours").agg({ "MathScore":"mean",	"ReadingScore":"mean",	"WritingScore":"mean"})
sns.heatmap(data=weekly_study_group,annot=True)
```

```python
df_melted=pd.melt(df,id_vars=["WklyStudyHours"],value_vars=["MathScore","ReadingScore","WritingScore"] ,var_name="WeeklyStudyHours",value_name="Scores")

plt.figure(figsize=(10, 6))  # Adjust figure size as needed
weekly_plots=sns.barplot(data=df_melted, x='WklyStudyHours', y='Scores', hue='WeeklyStudyHours')
weekly_plots.bar_label(weekly_plots.containers[0])
weekly_plots.bar_label(weekly_plots.containers[1])
weekly_plots.bar_label(weekly_plots.containers[2])
plt.title("Weekly Hours Distribution")
plt.show()
```

#### From Above analysis we Observer the Study Hours of students impacts the scores of Students

# Ethnic Group Distribution 

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()





```

# Matplotlib Ethnic Group Chart

```python
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]


labels = ['Group A', 'Group B', 'Group C', 'Group D', 'Group E']
 

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728',"#87CEEB"]

plt.figure(figsize=(8, 6))  # 
plt.pie(mlist, labels=labels, autopct='%1.2f%%',colors=colors,shadow=True)
plt.title("Ethnic Group Ditribution")
plt.show()
```
